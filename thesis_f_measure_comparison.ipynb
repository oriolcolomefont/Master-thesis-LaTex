```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.cm as cm

# Increase DPI in matplotlib's rcParams settings for the current Python session
plt.rcParams['figure.dpi'] = 1200

# Create dataframe
data = {'Author_Year': ['Turnbull et al., 2007', 'Kaiser et al., 2012', 'Sargent et al., 2011', 'McFee & Ellis, 2013',
                        'Nieto & Bello, 2014', 'Cannam et al., 2015', 'Nieto, 2016', 'Cannam et al., 2017',
                        'Ullrich et al., 2014', 'Grill & Schlter, 2015a', 'Grill & Schlter, 2015b',
                        'Hadria & Peeters, 2017', 'McCallum, 2019', 'Ours, 2023'],
        'Input': ['MFCCs, chromas, spectrogram', 'SSM', 'MFCCs, chromas', 'MLS',
                  'MFCCs, chromas', 'Timbre-type histograms', 'CQT Spectrogram',
                  'Timbre-type histograms', 'MLS', 'MLS, SSLMs', 'MLS, PCPs, SSLMs',
                  'MLS, SSLMs', 'CQT', 'Raw waveforms'],
        'Method': ['Boosted Decision Stump', 'Novelty measure', 'Viterbi',
                   'Fishers Linear Discriminant', 'Checkerboard-like kernel', 'HMM',
                   'Linear Discriminant Analysis', 'HMM', 'CNN', 'CNN', 'CNN', 'CNN',
                   'CNN', 'CNN'],
        'F_measure': [0.378, 0.286, 0.356, 0.317, 0.299, 0.213, 0.299, 0.212, 0.465,
                      0.523, 0.508, 0.291, 0.535, 0.288]}

df = pd.DataFrame(data)

# Normalize F_measure column for color mapping
normalize = plt.Normalize(df['F_measure'].min(), df['F_measure'].max())

# Create color map
cmap = cm.get_cmap("Reds")

# Create bar plot with color gradient
plt.figure(figsize=(10, 8))
plt.barh(df['Author_Year'], df['F_measure'], color=cmap(normalize(df['F_measure'].values)))
plt.xlabel('F-measure')
plt.ylabel('Author, Year')
plt.title('F-measure comparison of different unsupervised boundary detection studies')
plt.grid(True)
plt.colorbar(cm.ScalarMappable(norm=normalize, cmap=cmap))
```

```python
# Save as high resolution PNG
plt.savefig('f_measure_comparison.png', dpi=1200, bbox_inches='tight')
plt.show()
```
